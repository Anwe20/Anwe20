```python
#import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/HRDataset_v14.csv')
df.head()
```

```python
df.info()
```

```python
plt.hist(df['CitizenDesc'])
plt.title('Citizenship of Employees')
plt.xlabel('Type of Citizens')
plt.ylabel('Number of Employees')
```

```python
P=df[df['CitizenDesc']=='Eligible NonCitizen']
plt.barh(P['Employee_Name'],P['Salary'])
plt.title('Salary of Various Eligible Non Citizens')
```

```python
#heat map
fig = plt.figure(figsize=(8,5))
g = sns.heatmap(df.corr());
g.set_title('correlation between continuous variables [Heat Map]')
```

```python
#negative linear relationship
sns.lmplot(x='ManagerID', y='SpecialProjectsCount', data=df)
plt.title('Relationship Between Manager and Special Projects of Employee')
```

```python
#do abscences affect salary? No. Most of the salary is between 50000 and 100000 regardless of the number of times someone has been absent from work. In addition, the person with the highest salary has a 10.0 value for absences.

plt.scatter(df['Salary'], df['Absences'])
plt.xlabel('Salary')
plt.ylabel('Absences')
plt.title('Effect of Salary because of Absences')
plt.show()
```

```python
#Plot shows that the Software Engineering department has the highest satisfaction and Executive Office department has the lowest satisfaction for the employees.
sns.barplot(data=df, x='Department', y='EmpSatisfaction')
plt.xticks(rotation=45)
plt.title('Employee Satisfaction for Various Departments')
```

```python
sns.countplot(x=df['RecruitmentSource'])
plt.xticks(rotation=45)
plt.title('Recruiment Sources')
```

```python
sns.countplot(x=df['Sex'])
plt.title('Male vs Female Employees')
```

```python
sns.countplot(x=df['MaritalDesc'])
plt.title('Martial Status of Employees')
```

```python
fig = plt.figure(figsize=(20,5))

sns.countplot(x=df['Position'])
plt.title('Position of Employees in the Company')
plt.xticks(rotation=45)
```

```python
fig = plt.figure(figsize=(15,5))

sns.countplot(x=df['TermReason'])
plt.title('Termination Reason of Various Employees')
plt.xticks(rotation=45)
```

```python
sns.countplot(x=df['PerformanceScore'])
plt.title('Performance Score')
plt.xticks(rotation=45)
```
